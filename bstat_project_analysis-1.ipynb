```python
# Select the value we want to predict
predict = "G3"

# List the variables we want to use for our predictions in this model
data = data[[ "G1", "G2", "G3", "studytime", "health", "famrel", "failures", "absences","goout","internet"]]
data = shuffle(data)
# print(data)

x = np.array(data.drop([predict], 1))
y = np.array(data[predict])

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size = 0.15)


# training the model multiple times is needed to get high accuracy
best = 0
Accuracy_list = []
for _ in range(300):
    x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size = 0.15)

    linear = linear_model.LinearRegression()
#     linear = linear_model.LogisticRegression()

    linear.fit(x_train, y_train)
    acc = linear.score(x_test, y_test)
#     print("Accuracy: " + str(acc))
    
    Accuracy_list.append(acc)
    
    # We will be saving the best accuracy
    if (acc > best):
        best = acc
        with open("studentgrades.pickle", "wb") as f:
            pickle.dump(linear, f)
print("Highest Accuracy:", best)
avg = sum(Accuracy_list)/len(Accuracy_list)
print("The average accuracy is ", round(avg,2))

# Load model
pickle_in = open("studentgrades.pickle", "rb")
linear = pickle.load(pickle_in)

print("---------------------------------------------------------------------------")
print('Coefficient: \n', linear.coef_)
print('Intercept: \n', linear.intercept_)
print("---------------------------------------------------------------------------")

predictions = linear.predict(x_test)
# Print the predictions, the variables we used and the actual final grade
for x in range(len(predictions)):
    print("Predicted Final grade:", predictions[x], "Data:", x_test[x], "Actual Final grade:", y_test[x])

plot_data = {'Actual Final Grade':y_test,
        'Predicted Final Grade': predictions}
 
# Create DataFrame
df = pd.DataFrame(plot_data)

sns.regplot(x="Actual Final Grade", y="Predicted Final Grade", data=df);
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import seaborn as sn
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
data = pd.read_csv('students_math.csv')

# Select the value we want to predict
predict = "G3"

# Below are the vaiables which we are going to use for the prediction in our model
data = data[[ "G1", "G2", "G3", "studytime", "health", "famrel", "failures", "absences","goout","internet"]]
data = shuffle(data)
# print(data)

X = np.array(data.drop([predict], 1))
y = np.array(data[predict])

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.10,random_state=1)

logistic_regression= LogisticRegression()
logistic_regression.fit(X_train,y_train)
logistic_regression.score(X, y)

y_pred=logistic_regression.predict(X_test)

confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sn.heatmap(confusion_matrix, annot=True)

print('Accuracy: ',metrics.accuracy_score(y_test, y_pred))
plt.show()
```

```python
# R^2 and Absolute mean error
from sklearn.linear_model import Lasso, Ridge, ElasticNet, LinearRegression, LogisticRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
import warnings
warnings.filterwarnings('ignore')

# expand max number of display columns
pd.set_option('display.max_column', 100)

from matplotlib import pyplot as plt

# seaborn used for visualization
import seaborn as sns

data = pd.read_csv('students_math.csv')
data.shape
# remove duplicates data, if any... This is called data cleaning and also checking for null values is also called data cleaning
df = data.drop_duplicates()
df.isnull().sum()
print(df.shape)
predict = "G3"
data = data[["G1", "G2", "G3", "studytime", "health", "famrel", "failures", "absences", "goout", "internet"]]
data = shuffle(data)
# print(data)

X = np.array(data.drop([predict], 1))
y = np.array(data[predict])

# Training and test data Splitting using y and X
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

# make sure right number of data is available
print(len(X_train), len(X_test), len(y_train), len(y_test))
rg_lasso = Lasso(random_state=1234)
rg_ridge = Ridge(random_state=1234)
rf_logistic = LogisticRegression(random_state=1234)
rf = LinearRegression()
rg_en = ElasticNet(random_state=1234)
rg_rf = RandomForestRegressor(random_state=1234)
rg_gb = GradientBoostingRegressor(random_state=1234)
# create a dictionary to house all the instantiated models.
model_list = {'en': rg_en, 'rf': rg_rf, 'log': rf_logistic, 'lin': rf}

fitted_models = {}
for name, model in model_list.items():
    model.fit(X_train, y_train)

    fitted_models[name] = model

prediction = {}
for name, model in fitted_models.items():
    pred = model.predict(X_test)

    prediction[name] = pred
for name, pred in prediction.items():
    print(name)
    print('-------------------------------')
    print('R^2:', r2_score(y_test, pred))
    print('MAE:', mean_absolute_error(y_test, pred))
    print('\n')
```

```python
# Importing the libraries
# Random forest regression 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn import linear_model
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
from matplotlib import style
import pickle
import warnings
warnings.filterwarnings('ignore')


data = pd.read_csv('students_math.csv')

predict = "G3"
data = data[["G1", "G2", "G3", "studytime", "health", "famrel", "failures", "absences", "goout", "internet"]]
data = shuffle(data)

X = np.array(data.drop([predict], 1))
y = np.array(data[predict])
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.1)

best = 0
Accuracy_list = []
for _ in range(300):
    X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.25)

    rg_rf = RandomForestRegressor(random_state=1234)
    rg_rf.fit(X, y)
    acc = rg_rf.score(X_test, y_test)
    Accuracy_list.append(acc)
    if (acc > best):
        best = acc
        with open("sg.pickle", "wb") as f:
            pickle.dump(rg_rf, f)
print("Highest Accuracy:", best)
avg = sum(Accuracy_list) / len(Accuracy_list)
print("The average accuracy is ", round(avg, 2))

# Loading the model
pickle_in = open("sg.pickle", "rb")
randomForest = pickle.load(pickle_in)

predictions = randomForest.predict(X_test)
print(predictions)

for X in range(len(predictions)):
    print("Predicted Final grade:", predictions[X], "Data:", X_test[X], "Final grade:", y_test[X])
plot = "failures"
# Creating the visualization of the model
plt.scatter(data[plot], data["G3"], color="orange")
plt.legend("Comparision b/w failures Vs G3", loc=4)
plt.title("Comparision b/w failures vs G3(final grade)")
plt.xlabel("failures")
plt.ylabel("Final Grade")
plt.show()
```
